```python
films = []

# просматриваем необходимое количество страниц
for num_page in range(1):
    link_end = link + str(num_page)
    website_url = requests.get(link_end).text
    soup = BeautifulSoup(website_url, 'lxml')
    items = soup.find_all('div', class_ = 'row')

# собираем в список интересующую информацию  
    for item in items:
        films.append(
            {
            'film_name' : item.find('div', class_ = 'name-ru').get_text(strip=True),
            'series_name' :  item.find('div', class_ = 'alpha').get_text(strip=True),
            'part_name' :  item.find('div', class_ = 'left-part').get_text(strip=True),
            'release_date' :  item.find('div', class_ = 'right-part').get_text(strip=True),
            'series_raiting' :  item.find('div', class_ = 'mark-green-box').get_text(strip=True)
            })
        
# формируем датафрэйм
df = pd. DataFrame(films)
df
```

```python
# Создаем файл с результатом работы программы
df.to_csv('films.csv')
```
